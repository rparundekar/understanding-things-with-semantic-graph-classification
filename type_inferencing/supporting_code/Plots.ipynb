```python
def plotThese(title, toPlot, legendTitle,  legend):
    plt.clf()
    plt.cla()
    for array in toPlot:
        plt.plot(array)
        l = len(array)
    plt.ylim([0.5,1.0])
    plt.xlabel('Epochs')
    plt.ylabel('F1-score')
    plt.title(title)
    plt.xticks(range(0, l), range(1,l+1))
    plt.legend(legend, loc='upper left', title=legendTitle)
    plt.show()
```

```python
legend = ['Logistic','FC x 4','FC x 8']

toPlot = [attr_1xall_logistic, attr_1xall_full4, attr_1xall_full8]
plotThese('(a) Attribute Presence (# random walks = all)', toPlot, 'Network Structure', legend)

toPlot = [rel_1xall_logistic, rel_1xall_full4, rel_1xall_full8]
plotThese('(b) Relationship Presence (# random walks = all)', toPlot, 'Network Structure', legend)

toPlot = [inRel_1xall_logistic, inRel_1xall_full4, inRel_1xall_full8]
plotThese('(c) Incoming Relationship Presence (# random walks = all)', toPlot, 'Network Structure', legend)

toPlot = [all3_1xall_logistic, all3_1xall_full4, all3_1xall_full8]
plotThese('(d) All 3 (# random walks = all)', toPlot, 'Network Structure', legend)
```
